```python
## HW1- Vitor Melo
import numpy as np
import matplotlib.pyplot as plt
```

```python
def k_steadystate(A, theta, beta, delta):
    nom = A * theta
    denom = ((1 / beta) - (1 - delta))
    power = 1 / (1 - theta)
    answer = (nom / denom) ** power
    return answer
```

```python
def c_steadystate(A, k, theta, delta):
    y = A * (k ** theta)
    answer = y - delta * k
    return answer
```

```python

def k_cal_1(c, k, A, theta, delta):
    y = A * (k ** theta)
    answer = y + (1 - delta) * k - c
    return answer
```

```python
def c_cal_1(c, k, A, theta, delta, beta):
    y_k = theta * A * (k ** (theta - 1))
    answer = (y_k + (1 - delta)) * beta * c
    return answer
```

```python
def plot_klist(a, b):
    plt.plot(a, b, 'r')
    plt.xlabel('time')
    plt.ylabel('k_t')
    plt.title('k_t vs time')
    plt.show()


def plot_clist(a, b):
    plt.plot(a, b, 'b')
    plt.xlabel('time')
    plt.ylabel('c_t')
    plt.title('c_t vs time')
    plt.show()
```

```python

def list_k_c(c_start, k_start, A, theta, beta, delta):
    c_list = []
    k_list = []
    number = range(50)
    c_list.append(c_start)
    k_list.append(k_start)
    for n in number:
        k = k_cal_1(c_start, k_start, A, theta, delta)
        c = c_cal_1(c_start, k, A, theta, beta, delta)
        k_list.append(k)
        c_list.append(c)
        k_start = k
        c_start = c
    return k_list, c_list
```

```python

def final_iteration(A, theta, beta, delta, iteration, verbose=True):
    k_steady = k_steadystate(A, theta, beta, delta)
    print(k_steady)
    lowerbound = 0
    upperbound = k_steady
    k_0 = 0.5 * k_steady
    epsilon = 1e-1
    i = 0
    while True:
        c_0 = (lowerbound + upperbound) / 2
        y = A * (k_0 ** theta)
        if (c_0 > 0) and (c_0 > y):
            print("Is code coming to this place?")
            c_0 = c_0 - y - 0.90
        k_vec, c_vec = list_k_c(c_0, k_0, A, theta, beta, delta)
        error = k_vec[-1] - k_steady
        # check if the terminal condition is satisfied
        if abs(error) < epsilon:
            if verbose:
                print('Converged successfully on iteration ', i + 1)
                x_list = list(range(len(k_vec)))
                plot_klist(x_list, k_vec)
                plot_clist(x_list, c_vec)
            return k_vec, c_vec

        # if iteration continues, updates boundaries and guess of c_start
        if error < 0:
            upperbound = c_0
            print("Is code coming here, change upperbound?")
        else:
            lowerbound = c_0
            print("Is code coming here, change lowerbound?")
        i += 1
        if i == iteration:
            if verbose:
                print('Convergence failed.')
                x_list = list(range(len(k_vec)))
                plot_klist(x_list, k_vec)
                plot_clist(x_list, c_vec)
            return k_vec, c_vec


def main():
    answer = final_iteration(1, 0.4, 0.96, 0.1, 1000, True)
    print(answer)


if __name__ == "__main__":
    main()
```
